# Diagrama de Venn
Es un grfico sin ejes. Es un grfico compuesto por circulos que representan conjuntos de categoras. Los crculos se interesectan si existen observaciones que pertenecen a ambos conjuntos. El tamao de los crculos suele representar el nmero de elementos dentro de ese conjunto. Usualmente, podemos utilizar hasta un mximo de 3 conjuntos (3 crculos, 4 posibles intersecciones) 




### Cuando utilizar
- Quiero una forma visual de representar cuantas observaciones pertenecen al mismo tiempo a dos grupos diferentes. 


### Consejos
- Si tengo ms de 3 conjuntos el grfico se vuelve complicado de entender (ver imagen)
- No es comn utilizarlo en presentaciones, ni en exploracin de datos.

```python
from matplotlib_venn import venn2, venn2_unweighted
from matplotlib_venn import venn3, venn3_unweighted
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
df = pd.read_csv('./Datos/animales_caracteristicas.csv')
df
```

```python
respira = set(df[df['respira'] == 1]['animal'])
piernas = set(df[df['piernas'] == 1]['animal'])
nada = set(df[df['nada'] == 1]['animal'])
```

```python
# https://github.com/konstantint/matplotlib-venn
# https://towardsdatascience.com/how-to-create-and-customize-venn-diagrams-in-python-263555527305
venn3(
    [
        set(respira),
        set(piernas),
        set(nada)
    ],
    set_labels = ('Respira Aire', 'Tiene Piernas', 'Puede Nadar')
)
plt.title('Animales')
```

```python
venn3_unweighted(
    [
        set(respira),
        set(piernas),
        set(nada)
    ],
    set_labels = ('Respira Aire', 'Tiene Piernas', 'Puede Nadar')
)
plt.title('Animales')
```

```python
# Directamente definir las proporciones en los grupos
venn2(
    subsets=(30, 10, 5),
    set_labels=('Grupo A', 'Grupo B')
)
```
