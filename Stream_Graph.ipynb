# Stream Graph
Es una variacin del grfico de reas apiladas. El Stream Graph representa la variacion de magnitud de una variable en diferentes grupos a lo largo del tiempo. Tiene exactamente el mismo objetivo que el grfico de reas apiladas. Sin embargo, el Eje Y no es fijo. Es decir, no hay un "cero". Por lo que su objetivo no es demostrar presicin, si no cambios de proporcionalidad entre grupos. Incluso, es normal suavizar la lnea que une los puntos para que se vea mas agradable a la vista.    
**Eje X**: Variable categrica ordinal     
**Eje Y**: Variable numrica     
**Otros nombres**: Grfico de Flujo




### Cuando utilizar
- Comparar diferencias de proporcionalidad de propiedades numricas de diferentes grupos a lo largo del tiempo. 
- No me importa mucho exponer con presicin las magnitudes de mis datos.
- Quiero encontrar patrones temporales en mis datos.

### Cuando no utilizar
- Si quiero que las personas determinen con precisin la magnitud de mi variable numrica en cada grupo

### Consejos
- Rpidamente se puede convertir en un grfico difcil de interpretar. Por lo que su uso es limitado mayomermente a presentaciones visuales y no para demostrar presicin.
- Si tengo ms de 10 grupos debera pensar en un grfico de reas apiladas, o ser selectivo con mis categoras. O en su defecto, utilizar un generador de visualizaciones interactivo para quitar y mostrar categorias a gusto.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('./Datos/covid_data.csv') # Se lee el archivo csv
df # Se imprime el dataframe
```

```python
agrupado = df[['date', 'continent', 'new_cases']].groupby(['date', 'continent']).sum() # Se agrupa por fecha y continente
agrupado = agrupado.unstack(1) # Se desagrupa por continente
agrupado = agrupado.fillna(0) # Se rellenan los valores nulos con 0
agrupado = agrupado.droplevel(0, axis=1) # Se elimina el nivel 0 del multiindice
agrupado # Se imprime el dataframe
```

```python
# Streamgraph con Matplotlib
# https://matplotlib.org/stable/gallery/lines_bars_and_markers/stackplot_demo.html

values = [ # Se crea una lista con los valores de cada continente
    agrupado['Africa'], 
    agrupado['Asia'],
    agrupado['Europe'],
    agrupado['North America'],
    agrupado['Oceania'],
    agrupado['South America']
]
fig, ax = plt.subplots() # Se crea la figura y los ejes
fig.set_size_inches(10, 5) # Se establece el tamao de la figura
ax.stackplot( # Se crea el streamgraph
    agrupado.index.values, # Se establecen los valores del eje x
    values, # Se establecen los valores de los continentes
    labels=agrupado.columns.values, # Se establecen las etiquetas de los continentes
    baseline='wiggle', # Se establece el tipo de lnea base
    alpha=0.8 # Se establece la transparencia
)
ax.legend(loc='upper left') # Se aade la leyenda
ax.set_title('Casos de Covid por Continente')
ax.set_xlabel('Da')
ax.set_ylabel('Nmero de Casos')
ax.axhline(0, color="black", ls="--", linewidth=0.5) # Se aade una lnea horizontal
fig.autofmt_xdate() # Se ajustan las fechas
fig.tight_layout() # Se ajusta el layout
plt.xticks(np.arange(0, len(agrupado.index.values), 30)) # Se establece la separacin de los ticks
plt.show()
```

```python
# Con curva suavizada por una funcion gaussiana
# https://www.python-graph-gallery.com/streamchart-basic-matplotlib

import scipy.stats as stats # Se importa la librera stats de scipy

def gaussian_smooth(x, y, grid, sd): # Se define la funcin para suavizar
    weights = np.transpose([stats.norm.pdf(grid, m, sd) for m in x]) # Se calculan los pesos
    weights = weights / weights.sum(0) # Se normalizan los pesos
    return (weights * y).sum(1) # Se retorna el suavizado

x = np.arange(0, len(agrupado.index.values)) # Se establece el eje x
grid = np.linspace(0, len(agrupado.index.values), num=len(agrupado.index.values)) # Se establece la malla

y_suavizado = [gaussian_smooth(x, list(y_), grid, 1) for y_ in values] # Se suavizan los valores
```

```python
fig, ax = plt.subplots() # Se crea la figura y los ejes
fig.set_size_inches(15, 10) # Se establece el tamao de la figura
ax.stackplot( # Se crea el streamgraph
    agrupado.index.values, 
    y_suavizado, 
    labels=agrupado.columns.values, 
    baseline='wiggle',
    alpha=0.8
)
ax.legend(loc='upper left') # Se aade la leyenda
ax.set_title('Casos de Covid por Continente') # Se aade el ttulo
ax.set_xlabel('Da') # Se aade el ttulo del eje x
ax.set_ylabel('Nmero de Casos') # Se aade el ttulo del eje y
ax.axhline(0, color="black", ls="--", linewidth=0.5) # Se aade una lnea horizontal
fig.autofmt_xdate() # Se ajustan las fechas
fig.tight_layout() # Se ajusta el layout
plt.xticks(np.arange(0, len(agrupado.index.values), 30)) # Se establece la separacin de los ticks
plt.show()
```
